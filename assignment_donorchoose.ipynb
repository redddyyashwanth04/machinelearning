```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

# Get English stopwords
stop_words = set(stopwords.words('english'))
print(stop_words)
```

```python

import re
def decontracted(phrase):
    # specific
    phrase = re.sub(r"won't", "will not", phrase)
    phrase = re.sub(r"can\'t", "can not", phrase)
    phrase = re.sub(r"\'s", " is", phrase)
    phrase = re.sub(r"\'d", " would", phrase)
    phrase = re.sub(r"\'ll", " will", phrase)
    phrase = re.sub(r"\'t", " not", phrase)
    phrase = re.sub(r"\'ve", " have", phrase)
    phrase = re.sub(r"\'m", " am", phrase)
    return phrase
```

```python
from tqdm import tqdm

def preprocess_text(text_data):
  preprocessed_text=[]
  for sentance in tqdm(text_data):
    sentance=re.sub(r"http\S+", "", sentance)
    sentance=decontracted(sentance)
    sentance=re.sub("\S*\d\S*", "", sentance).strip()
    sentance=re.sub('[^A-Za-z]+', ' ', sentance)
    sentance = sentance.replace('\\r', ' ')
    sentance = sentance.replace('\\n', ' ')
    sentance = sentance.replace('\\"', ' ')
    sentance = re.sub('[^A-Za-z0-9]+', ' ', sentance)
    sentance = ' '.join(e.lower() for e in sentance.split() if e.lower() not in stop_words)
    preprocessed_text.append(sentance.strip())
  return preprocessed_text
# training_data["project_title"]=preprocess_text(training_data["project_title"])
print(training_data["project_title"].head())
preprocessed_titles=preprocess_text(training_data['project_title'])

preprocessed_titles[:5]#this is the list and you can only print the elements in the list if you want print the elements or txt.
#you cant access the list like this training_data['project_title']
print(type(preprocessed_titles))
```

```python

from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import roc_auc_score
model_set_1 = MultinomialNB()
param_grid = {'alpha':[0.0001,0.0005,0.001,0.005,0.01,0.05,0.1,1, 2,5,100,500]}
clf_set_1 = GridSearchCV(model_set_1,param_grid, cv= 3, scoring='roc_auc',return_train_score=True)
clf_set_1.fit(X_train_final, y_train)
train_auc= clf_set_1.cv_results_['mean_train_score']
cv_auc=clf_set_1.cv_results_['mean_test_score']
```

```python
def batch_predict(clf, data):
    # roc_auc_score(y_true, y_score) the 2nd parameter should be probability estimates of the positive class
    # not the predicted outputsy_data_pred = []
    y_data_pred = []
    tr_loop = data.shape[0] - data.shape[0]%1000
    # consider you X_tr shape is 49041, then your tr_loop will be 49041 - 49041%1000 = 49000
    # in this for loop we will iterate unti the last 1000 multiplier
    for i in range(0, tr_loop, 1000):
        y_data_pred.extend(clf.predict_proba(data[i:i+1000])[:,1])
    # we will be predicting for the last data points
    if data.shape[0]%1000 !=0:
        y_data_pred.extend(clf.predict_proba(data[tr_loop:])[:,1])

    return y_data_pred
```

```python
from sklearn.metrics import roc_curve, auc
model_set_1_best = MultinomialNB(alpha=best_params['alpha'])
model_set_1_best.fit(X_train_final, y_train)
y_train_pred = batch_predict(model_set_1_best, X_train_final)
y_test_pred = batch_predict(model_set_1_best, X_test_final)
train_fpr, train_tpr, tr_thresholds = roc_curve(y_train, y_train_pred)
test_fpr, test_tpr, te_thresholds = roc_curve(y_test, y_test_pred)
plt.plot(train_fpr, train_tpr, label="train AUC ="+str(auc(train_fpr, train_tpr)))
plt.plot(test_fpr, test_tpr, label="test AUC ="+str(auc(test_fpr, test_tpr)))
plt.legend()
plt.ylabel("True Positive Rate -> TPR")
plt.xlabel("False Positive Rate -> FPR")
plt.title("ROC-AUC curve")
plt.grid()
```
